# train, test 셋 나누기


x_train, x_valid, y_train, y_valid = train_test_split(data, target, test_size=0.2, shuffle=True, stratify=target, random_state=34)

1. test_size - 0.2는 전체 데이터 셋의 20%를 test (validation) 셋으로 지정.default는 0.25

2. shuffle -  split을 해주기 이전에 섞을건지 여부, default = True

3. stratify - classification을 다룰 때 매우 중요. stratify 값을 target으로 지정해주면 각각의 class 비율(ratio)을 train / validation에 유지
즉, 한 쪽에 쏠려서 분배되는 것을 방지

4. random_state - 이 값을 고정해두고 튜닝해야 매번 데이터셋이 변경되는 것을 방지

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, 
                                                    test_size=0.3, 
                                                    random_state=0, stratify = target)
```

# XGBoost 란?!?!

- Extreme Gradient Boosting의 약자
- 여러 개의 약한 의사결정나무(Decision Tree)를 조합해서 사용하는 앙상블(Ensemble) 기법
- Regression, Classification 문제를 모두 지원

## 장점?!
- 과적합 규제(Regularization)
- Missing Values : 결측치를 내부적으로 처리
- CART(Classification and regression tree) 앙상블 모델 

## parameter
- n_estimators = The number of trees (or rounds), default = 100 
- learning - rate =  how fast the model learns.

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimators=500, learning_rate=0.05, random_state=100)

# 성능 평가 지표를 auc로, 조기 중단 파라미터는 100으로 설정하고 학습 수행. 
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_train, y_train), (X_test, y_test)])
```
